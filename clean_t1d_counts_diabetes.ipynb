```python
import pandas as pd
import numpy as np

data=pd.read_csv('/content/clean_t1d_counts.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.isnull().values.any()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
data['characteristics']=label.fit_transform(data['characteristics'])
data.head()
```

```python
x=data.drop(['GSM','characteristics'],axis=1)
y=data['characteristics']
print(x)
print(y)
```

```python
'''from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
std_data=scaler.transform(x)
print(std_data)'''
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,stratify=y,random_state=2)
print(x.shape,x_train.shape,x_test.shape)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train,y_train)
y_train_pred=rf.predict(x_train)
rf_train_acc=accuracy_score(y_train_pred,y_train)
print(rf_train_acc)

y_test_pred=rf.predict(x_test)
rf_test_acc=accuracy_score(y_test_pred,y_test)
print(rf_test_acc)
```

Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

gb=GradientBoostingClassifier(random_state=42)
gb.fit(x_train,y_train)
y_train_pred=gb.predict(x_train)
gb_train_acc=accuracy_score(y_train_pred,y_train)
print(gb_train_acc)

y_test_pred=gb.predict(x_test)
gb_test_acc=accuracy_score(y_test_pred,y_test)
print(gb_test_acc)
```

XGBoosting

```python
from xgboost import XGBClassifier

xgb=XGBClassifier(random_state=42)
xgb.fit(x_train,y_train)
y_train_pred=xgb.predict(x_train)
xgb_train_acc=accuracy_score(y_train_pred,y_train)
print(xgb_train_acc)

y_test_pred=xgb.predict(x_test)
xgb_test_acc=accuracy_score(y_test_pred,y_test)
print(xgb_test_acc)
```

CATBoost

```python
from catboost import CatBoostClassifier

cat=CatBoostClassifier(random_state=42)
cat.fit(x_train,y_train)
y_train_pred=cat.predict(x_train)
cat_train_acc=accuracy_score(y_train_pred,y_train)
print(cat_train_acc)

y_test_pred=cat.predict(x_test)
cat_test_acc=accuracy_score(y_test_pred,y_test)
print(cat_test_acc)
```

```python
from sklearn.ensemble import VotingClassifier
ensemble=VotingClassifier(estimators=[('rf',rf),('gb',gb),('cat',cat),('xgb',xgb)],voting='soft')
ensemble.fit(x_train,y_train)
y_train_pred=ensemble.predict(x_train)
ensemble_train_acc=accuracy_score(y_train_pred,y_train)
print(ensemble_train_acc)

y_test_pred=ensemble.predict(x_test)
ensemble_test_acc=accuracy_score(y_test_pred,y_test)
print(ensemble_test_acc)
```

Ada boost

```python
from sklearn.ensemble import AdaBoostClassifier
ada=AdaBoostClassifier(random_state=42)
ada.fit(x_train,y_train)
y_train_pred=ada.predict(x_train)
ada_train_acc=accuracy_score(y_train_pred,y_train)
print(ada_train_acc)

y_test_pred=ada.predict(x_test)
ada_test_acc=accuracy_score(y_test_pred,y_test)
print(ada_test_acc)
```

SVM

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train_scaled=scaler.transform(x_train)
print(x_train_scaled)
scaler.fit(x_test)
x_test_scaled=scaler.transform(x_test)
print(x_test_scaled)
```

```python
from sklearn import svm
svc=svm.SVC(kernel='linear')
svc.fit(x_train,y_train)
y_train_pred=svc.predict(x_train)
svc_train_acc=accuracy_score(y_train_pred,y_train)
print(svc_train_acc)

y_test_pred=svc.predict(x_test)
svc_test_acc=accuracy_score(y_test_pred,y_test)
print(svc_test_acc)
```

```python
result=pd.DataFrame({'Models':['Random Forest','Gradient Boosting', 'XGBoost', 'CatBoost', 'AdaBoost', 'Voting Ensemble', 'SVM'],
                     'Train Accuracy':[rf_train_acc, gb_train_acc, xgb_train_acc, cat_train_acc, ada_train_acc, ensemble_train_acc, svc_train_acc],
                     'Test Accuracy':[rf_test_acc, gb_test_acc, xgb_test_acc, cat_test_acc, ada_test_acc, ensemble_test_acc, svc_test_acc]})
print(result)
```
